```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
df = pd.read_csv("House_Prices _Full.csv")
df
```

```python
df.drop(['date','street','date', 'sqft_living', 'waterfront','city','country','statezip'],axis = 1,inplace= True)
df
```

```python
df.shape
```

```python
df['yr_built'] = 2014 - df['yr_built']
```

```python
df['yr_renovated'] = df['yr_renovated'].apply(lambda x: 0 if (x == 0) else 2014 - x)
df
```

```python
X = df.drop(['sqft_lot','floors','yr_renovated','price'],axis =1)
y = df['price']

X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15,random_state= 42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_scale = ss.fit_transform(X_train)
X_test_scale = ss.transform(X_test)
```

```python
linearRegression = linear_model.LinearRegression()
```

```python
linearRegression.fit(X_train_scale,y_train)
```

```python
y_pred_linearRegression = linearRegression.predict(X_test_scale)
```

```python
rscore_linearRegression = r2_score(y_test,y_pred_linearRegression)
```

```python
print('R2 Linear Regression: {}'.format(rscore_linearRegression))
```

```python

plt.plot(y_pred_linearRegression,y_pred_linearRegression,c='r')
plt.scatter(y_test,y_test,c='g')
plt.title('Linear Regression')
plt.legend(['y_pred_linearRegression','y_pred'])
```
